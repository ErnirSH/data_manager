```python
import pandas as pd
```

```python
data = pd.read_csv('../data/BTCUSDT.csv', parse_dates=['timestamp'])
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.dtypes
```

```python
df[df['volume'] == 0]
```

```python
df['is_missing'] = df['volume'] == 0
```

```python
df['change'] = df['is_missing'] != df['is_missing'].shift()
```

```python
df['group'] = df['change'].cumsum()
```

```python
grouped = df.groupby('group')
```

```python
group_sizes = grouped.size().reset_index(name='count')
```

```python
group_sizes['timestamp'] = grouped['timestamp']
```
